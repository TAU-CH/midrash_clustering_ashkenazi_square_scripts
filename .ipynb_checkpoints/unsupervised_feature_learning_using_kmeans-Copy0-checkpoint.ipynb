```python
import os
import random
from PIL import Image

# Define the folder containing the image folders.
folder_path = "grayscale_text_regions"
# Define the patch size.
patch_size = (16, 16)
# Define the total number of patches you want to sample from each folder.
n = 10

# Initialize a list to store the patches.
patches = []

```

```python
# Step 1: Loop through each folder and sample 'n' patches based on image sizes.
for folder_name in os.listdir(folder_path):
    folder = os.path.join(folder_path, folder_name)
    if os.path.isdir(folder):
        # Get a list of image files in the folder.
        image_files = [os.path.join(folder, file) for file in os.listdir(folder) if file.endswith(".png")]

        # Calculate the weight for each image based on its height x width pixel size.
        total_size = sum(Image.open(file).size[0] * Image.open(file).size[1] for file in image_files)
        image_weights = [Image.open(file).size[0] * Image.open(file).size[1] / total_size for file in image_files]

        # Distribute 'n' patches based on image weights within the folder.
        for _ in range(n):
            chosen_image = random.choices(image_files, weights=image_weights)[0]
            image = Image.open(chosen_image)
            width, height = image.size
            x = random.randint(0, width - patch_size[0])
            y = random.randint(0, height - patch_size[1])
            patch = image.crop((x, y, x + patch_size[0], y + patch_size[1]))
            patches.append(patch)

# Now, 'patches' contains 'mxn' patches, with 'n' patches sampled from each of the 'm' folders based on image sizes.
```

```python
import numpy as np

# Convert the patches to NumPy array for K-Means
patch_array = np.array([np.array(patch) for patch in patches]).reshape(-1, patch_size[0] * patch_size[1])


def normalize_for_contrast(patch_array):
    """ Normalize for contrast """
    patch_mean = np.mean(patch_array, axis=1, keepdims=True)
    patch_variance = np.var(patch_array, axis=1, ddof=1, keepdims=True)
    normalized_patches = (patch_array - patch_mean) / (np.sqrt(patch_variance + 10))
    return normalized_patches

def data_whitening(patch_array):
    """ Whitening """
    C = np.cov(patch_array, rowvar=False)
    M = np.mean(patch_array, axis=0)
    d, V = np.linalg.eig(C)
    D = np.diag(np.sqrt(1. / (d + 0.1)))
    P = np.dot(np.dot(V, D), V.T)
    whitened_patches = np.dot(patch_array - M, P)
    return whitened_patches

# Normalize patches for contrast
normalized_patches = normalize_for_contrast(patch_array)

# Whitening
whitened_patches = data_whitening(normalized_patches)

# 'whitened_patches' now contains the whitened patches

```

```python
import random
import matplotlib.pyplot as plt

# Number of patches to visualize
num_patches_to_visualize = 80

# Select random indices
selected_indices = random.sample(range(len(patch_array)), num_patches_to_visualize)

# Extract the patches from each dataset
selected_patch_array = patch_array[selected_indices]
selected_normalized_patches = normalized_patches[selected_indices]
selected_whitened_patches = whitened_patches[selected_indices]

height = 16
width = 16

# Reshape the patches to (num_patches_to_visualize, height, width)
reshaped_original_patches = selected_patch_array.reshape(num_patches_to_visualize, height, width)
reshaped_normalized_patches = selected_normalized_patches.reshape(num_patches_to_visualize, height, width)
reshaped_whitened_patches = selected_whitened_patches.reshape(num_patches_to_visualize, height, width)

# Create three separate figures to visualize the three groups of patches

# Figure 1: Original Patches
plt.figure(figsize=(12, 4))
plt.suptitle("Original Patches")
for i in range(num_patches_to_visualize):
    plt.subplot(8, 10, i + 1)
    plt.imshow(reshaped_original_patches[i], cmap='gray')
    plt.axis('off')

# Figure 2: Normalized Patches
plt.figure(figsize=(12, 4))
plt.suptitle("Normalized Patches")
for i in range(num_patches_to_visualize):
    plt.subplot(8, 10, i + 1)
    plt.imshow(reshaped_normalized_patches[i], cmap='gray')
    plt.axis('off')

# Figure 3: Whitened Patches
plt.figure(figsize=(12, 4))
plt.suptitle("Whitened Patches")
for i in range(num_patches_to_visualize):
    plt.subplot(8, 10, i + 1)
    plt.imshow(reshaped_whitened_patches[i], cmap='gray')
    plt.axis('off')

plt.show()
```

```python
from sklearn.cluster import KMeans

# Assuming you have 'whitened_patches' containing the whitened patches
number_of_centroids = 10  # Set the number of centroids

# Initialize the K-means model with the chosen number of centroids
kmeans = KMeans(n_clusters=number_of_centroids, random_state=0)

# Fit the K-means model to your data (whitened_patches)
kmeans.fit(whitened_patches)

# Get the learned centroids
centroids = kmeans.cluster_centers_
```

```python
def hard_assignment_coding(centroids, patches):
    """
    Apply hard-assignment coding to a set of patches using learned centroids.
    
    Parameters:
        centroids (numpy.ndarray): The learned centroids.
        patches (numpy.ndarray): The set of patches to be encoded.
    
    Returns:
        numpy.ndarray: The hard-assignment coding for each patch.
    """
    num_patches = patches.shape[0]
    num_centroids = centroids.shape[0]
    coding = np.zeros((num_patches, num_centroids), dtype=int)

    for i in range(num_patches):
        distances = np.linalg.norm(centroids - patches[i], axis=1)
        min_index = np.argmin(distances)
        coding[i, min_index] = 1

    return coding
```

```python
import os
import random
import numpy as np
from PIL import Image

# Define the folder containing the text region images
folder_path = 'grayscale_text_regions'

# Number of patches to sample per document
n_patches_per_document = 100

# Patch size
patch_size = (16, 16)

# List to store codings for each document
document_codings = []


# Step 1: Loop through each folder (document page).
for folder_name in os.listdir(folder_path):
    folder = os.path.join(folder_path, folder_name)
    if os.path.isdir(folder):
        # Get a list of image files in the folder.
        image_files = [os.path.join(folder, file) for file in os.listdir(folder) if file.endswith(".png")]

        # Calculate the weight for each image based on its height x width pixel size.
        total_size = sum(Image.open(file).size[0] * Image.open(file).size[1] for file in image_files)
        image_weights = [Image.open(file).size[0] * Image.open(file).size[1] / total_size for file in image_files]

        # List to store patches for this document
        document_patches = []

        # Step 2: Distribute patches based on image weights within the folder.
        for _ in range(n_patches_per_document):
            chosen_image = random.choices(image_files, weights=image_weights)[0]
            image = Image.open(chosen_image)
            width, height = image.size
            x = random.randint(0, width - patch_size[0])
            y = random.randint(0, height - patch_size[1])
            patch = image.crop((x, y, x + patch_size[0], y + patch_size[1]))
            document_patches.append(patch)

        # Convert patches to a numpy array
        patches_array = np.array([np.array(patch) for patch in document_patches])

       # Step 3: Normalize patches for contrast
        normalized_patches = normalize_for_contrast(patches_array)

        # Step 4: Apply data whitening
        whitened_patches = data_whitening(normalized_patches)

        # Step 5: Apply hard_assignment_coding to get codings for the patches in this document.
        centroids = np.array([centroids])  # You should have your centroids here
        codings = hard_assignment_coding(centroids, whitened_patches)

        # Append the codings for this document
        document_codings.append(codings)

# Now you have a list of codings for each document in document_codings.
```
